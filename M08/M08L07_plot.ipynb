# Jak najszybciej umieścić dane na wykresie?

```python
import pandas as pd

df = pd.read_csv(
    'data/forex_usd_data.csv',
    parse_dates=['date(y-m-d)'],
    index_col='date(y-m-d)',
).rename(columns={
    'Polish Zloty': 'PLN',
    'Euro': 'EUR',
})

# Do wykresów w Pythonie służy m.in. biblioteka matplotlib.
# Pandas posiada już nakładkę na tą bibliotekę, tak że można łatwo wywołać metodę .plot():

df.plot(
    y='PLN',  # nazwa kolumny (lub lista kolumn) do umieszczenia na osi Y. Na osi X będziemy mieć index.
    figsize=(15, 5),  # wymiary wykresu w calach
)  # ✅ średnik zapobiega wyświetlaniu tego, co zwraca funkcja plot() - dzięki temu nie widzimy `<AxesSubplot:xlabel='date(y-m-d)'>`
```

```python
import pandas as pd
import matplotlib.pyplot

macro = pd.read_csv (
        'data\macroeconomicsPL.csv',
        parse_dates=['DATE'],
        index_col='DATE',
        dayfirst=True,
        skipfooter=3,
        engine="python",
        thousands=',',
        na_values={
            'SPD': '-99'
        },
).rename(columns={
    'sr_NBP': 'SR',
    'WIBOR 3M': 'WIBOR'
})

macro.plot(
    y=['SR', 'CPI'],
    figsize=(10,5),
    grid=True
);
```
